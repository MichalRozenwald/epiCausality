```python
def plot_fracs(fracs_dict, title="Methylation Fractions",
               figsize=(16, 5), alpha=0.6, colors=None):
    """Bar plot comparing methylation fractions across conditions.

    Parameters
    ----------
    fracs_dict : dict[str -> np.ndarray]
        {condition_label: fracs_array}  — plotted in insertion order.
    title : str
    figsize : tuple
    alpha : float
    colors : list of str, optional
    """
    default_colors = ["deepskyblue", "navy", "tomato", "darkred",
                      "mediumseagreen", "darkgreen"]
    colors = colors or default_colors
    n_cpgs = next(iter(fracs_dict.values())).shape[0]
    x = np.arange(n_cpgs)

    fig, ax = plt.subplots(figsize=figsize)
    for i, (label, fracs) in enumerate(fracs_dict.items()):
        ax.bar(x, fracs, snap=False, alpha=alpha,
               label=label, color=colors[i % len(colors)])
    ax.set_xlabel("CpG Index")
    ax.set_ylabel("Methylation Fraction")
    ax.set_title(title)


    # Set x-axis ticks to show all CpG indices
    ax.set_xticks(np.arange(n_cpgs))
    ax.set_xticklabels(np.arange(n_cpgs), rotation=90, fontsize=8)

    ax.legend()
    plt.tight_layout()
    plt.show()


# Day 6: CRoff vs Unedited, mc0.7
plot_fracs(
    {"CRISPRoff (mc>0.995)": filtered_CGs_D6_CRoff_mc0995_fracs,
     "Unedited  (mc>0.995)": filtered_CGs_D6_unedited_mc0995_fracs},
    title="Day 6 — CD55 CpG Methylation Fractions (Filtered, mC threshold 0.995)",
    colors=["skyblue", "royalblue"])
```

```python
import numpy as np
"""
Scale dif_mc to integer scores in 0..1000 suitable for UCSC BED 'score'.
Steps:
    1. compute min (mn) and max (mx) of dif_mc
    2. if mx == mn -> return constant mid score (500) array
    3. otherwise: (dif_mc - mn) / (mx - mn) * 1000
    4. round, clip to [0,1000], replace NaN with 0, cast to int
 """

# scale dif_mc to range 0..1000
mn = np.nanmin(dif_mc)
mx = np.nanmax(dif_mc)

if np.isclose(mx, mn):
    dif_mc_scaled = np.full_like(dif_mc, 500.0)  # constant -> mid score
else:
    dif_mc_scaled = (dif_mc - mn) / (mx - mn) * 1000.0

# round, clip to [0,1000], replace NaN with 0 and convert to int
dif_mc_scaled_round = np.round(dif_mc_scaled)
dif_mc_scaled_clip = np.clip(dif_mc_scaled_round, 0, 1000)
dif_mc_scaled_clip_nan0 = np.where(np.isnan(dif_mc_scaled_clip), 0, dif_mc_scaled_clip).astype(int)

# assign back or use new variable
dif_mc_scaled_int = dif_mc_scaled_clip_nan0

print("orig_min, orig_max:", mn, mx)
print("scaled_min, scaled_max:", dif_mc_scaled_int.min(), dif_mc_scaled_int.max())
```

```python
def make_track_difs_color_bed(dif_mc: np.ndarray, 
                              CG_coordinates: np.ndarray,
                                    experiment_condition: str, 
                                    out_path: Path) -> None:
    """
    BED9 with itemRgb, colors reflect dif_mc values:

    - sign:   red = positive, blue = negative
    - brightness: ~ |effect_size| (robustly scaled by 95th percentile)
    """

    es = dif_mc.astype(float)

    # robust scaling by |effect_size|
    abs_es = np.abs(np.where(np.isinf(es), np.nan, es))
    if np.isfinite(abs_es).any():
        p95 = np.nanpercentile(abs_es[np.isfinite(abs_es)], 95)
    else:
        p95 = 1.0
    if p95 == 0 or np.isnan(p95):
        p95 = 1.0

    norm = np.clip(abs_es / p95, 0, 1)

    # intensity range (bright)
    min_intensity = 150
    intensity = (min_intensity + norm * (255 - min_intensity)).round().astype(int)

    r = np.zeros(len(es), dtype=int)
    g = np.zeros(len(es), dtype=int)
    b = np.zeros(len(es), dtype=int)

    pos_mask = es > 0
    neg_mask = es < 0
    zero_mask = ~(pos_mask | neg_mask)

    # positive -> shades of red
    r[pos_mask] = intensity[pos_mask]
    g[pos_mask] = 0
    b[pos_mask] = 0

    # negative -> shades of blue
    r[neg_mask] = 0
    g[neg_mask] = 0
    b[neg_mask] = intensity[neg_mask]

    # zeros/NaNs -> neutral gray
    r[zero_mask] = 180
    g[zero_mask] = 180
    b[zero_mask] = 180

    itemRgb = [f"{ri},{gi},{bi}" for ri, gi, bi in zip(r, g, b)]

    start_coords = CG_coordinates.astype(int)
    end_coords = CG_coordinates.astype(int) + 2

    bed_score = (norm * 1000).round().astype(int)

    bed9 = pd.DataFrame(
        {
            "chrom": "chr1",
            "start": CG_coordinates,
            "end": np.array(CG_coordinates) + 2,
            "name": "CpG_" + np.arange(1, len(CG_coordinates) + 1).astype(str),
            "score": bed_score,
            "strand": "+",
            "thickStart": start_coords,
            "thickEnd": end_coords,
            "itemRgb": itemRgb,
        }
    )
    track_line = (
        'track name="dif_mCG_Red_Blue_'+experiment_condition+'" '
        'description="CG colored RB by dif_mCG (red=positive, blue=negative; brightness~|dif_mCG|) '+experiment_condition+'" '
        'visibility=2 itemRgb="On"\n'
    )

    with open(out_path, "w") as f:
        f.write(track_line)
    bed9.to_csv(out_path, sep="\t", header=False, index=False, mode="a")
    print(f"Written effect_size-colored BED to: {out_path}")
```
