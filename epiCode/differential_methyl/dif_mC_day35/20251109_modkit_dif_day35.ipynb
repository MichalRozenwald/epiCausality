Preparing the input data
The inputs to all modkit dmr commands are two or more bedMethyl files (created by modkit pileup) that have been compressed with bgzip and indexed with tabix. An example of how to generate the input data is shown below:


ref=grch38.fasta
threads=32

norm=normal_sample.bam
norm_pileup=normal_pileup.bed

modkit pileup ${norm} ${norm_pileup} \
  --cpg \
  --ref ${ref} \
  --threads ${threads} \
  --log-filepath log.txt

bgzip -k ${norm_pileup}
tabix -p bed ${norm_pileup}.gz

# pileup and compression can also be done in one step
tumor=tumor_sample.bam
tumor_pileup=tumor_pileup.bed.gz

modkit pileup ${tumor} - \
  --cpg \
  --ref ${ref} \
  --threads ${threads} \
  --log-filepath log.txt | ${bgzip} -c > ${tumor_pileup}

tabix -p bed ${tumor_pileup}

3. Detecting differential modification at single base positions
The modkit dmr pair command has the ability to score individual bases (e.g. differentially methylated CpGs). To run single-base analysis on one or more paired samples, simply omit the --regions (-r) option when running modkit dmr pair. When performing single-base analysis the likelihood ratio score and a MAP-based p-value are available. For details on the likelihood ratio score and the MAP-based p-value, see the scoring details section. For example the above command becomes:

dmr_result=single_base_haplotype_dmr.bed

modkit dmr pair \
  -a ${hp1_pileup}.gz \
  -b ${hp2_pileup}.gz \
  -o ${dmr_result} \
  --ref ${ref} \
  --base C \
  --threads ${threads} \
  --log-filepath dmr.log

```python
# Visualize modkit DMR result
import os
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go

# Path to the DMR BED output (adjust if you use a different file)
dmr_path = "/home/michalula/code/epiCausality/epiCode/differential_methyl/modkit_output/single_base_day28_CRoff_vs_Unedit_dmr_20251109.bed"
out_dir = os.path.dirname(dmr_path) or "."
os.makedirs(out_dir, exist_ok=True)

# Try reading with a header first, fall back to headerless read
try:
    df = pd.read_csv(dmr_path, sep='	', comment='#', header=0, engine='python')
    if df.shape[1] < 3:
        df = pd.read_csv(dmr_path, sep='	', comment='#', header=None, engine='python')
except Exception as e:
    print("Failed to read DMR file:", e)
    raise

# If the file had no header, pandas will give integer column names - assign common BED/DMR names
if isinstance(df.columns[0], int):
    ncols = df.shape[1]
    default_names = ["chrom", "start", "end", "name", "score", "strand"]
    names = default_names[:ncols] + [f"col{i}" for i in range(len(default_names), ncols)]
    df.columns = names

# Coerce numeric-looking columns where possible
for c in df.columns:
    try:
        df[c] = pd.to_numeric(df[c], errors='ignore')
    except Exception:
        pass

print("DMR file:", dmr_path)
print("Rows,cols:", df.shape)
display(df.head())
display(df.describe(include="all").T)

# Make histograms of numeric columns and save as HTML
numcols = df.select_dtypes(include=[np.number]).columns.tolist()
if len(numcols):
    for col in numcols:
        fig = px.histogram(df, x=col, nbins=80, title=f"Histogram of {col}")
        html = os.path.join(out_dir, f"dmr_hist_{col}.html")
        fig.write_html(html)
        print("Saved", html)

# Scatter: genomic position vs score (if present)
score_col = None
if 'start' in df.columns:
    if 'score' in df.columns:
        score_col = 'score'
    else:
        # try to find a numeric column to plot against position
        for c in numcols:
            if c != 'start' and c != 'end':
                score_col = c
                break
    if score_col is not None:
        df_plot = df.copy()
        df_plot['pos'] = df_plot['start'].astype(int)
        fig = px.scatter(df_plot, x='pos', y=score_col, color='chrom' if 'chrom' in df_plot.columns else None,
                         title=f"Position vs {score_col}", hover_data=df_plot.columns.tolist(), height=600)
        html = os.path.join(out_dir, 'dmr_position_score_scatter.html')
        fig.write_html(html)
        print("Saved", html)

# Show a table of top hits (by p-value or by score if p-value not present)
rank_col = None
for key in ['pvalue','p_val','p.value','p','padj','qval','score']:
    if key in df.columns:
        rank_col = key
        break
if rank_col is None and len(numcols):
    rank_col = numcols[0]
if rank_col is not None:
    ascending = True if 'p' in str(rank_col).lower() or 'q' in str(rank_col).lower() else False
    top = df.sort_values(by=rank_col, ascending=ascending).head(50)
    display(top)

print("Done. Visualizations saved to:", out_dir)
```
