```python
C_CG_coords_fwd=np.array(C_CG_coords)

def extract_mC_probs_from_BS_bedgraph(bedgraph_path, is_forward_strand,
                                      C_CG_coords_fwd=[207320868,207320873,207320878,207320880,207320882,207320884,207320886,207320888,207320890,207320892]
                                      ):
  # For the data from the Reverese strand: Shift by one the coordinates of the C locations
  if is_forward_strand == False:
    C_CG_coords_fwd += 1
    print("This file is from the Reverse strand, so shifting the C coordinates by one")
    print(C_CG_coords_fwd)

  # Read the BedGraph File: Use Python's built-in file handling methods to read the .bedgraph file. Here’s how you might start:
  with open(bedgraph_path, 'r') as file:
      bedgraph_data = file.readlines()

  # Process Data with Pandas: Import the Pandas library to manage and analyze your data effectively. You can convert your data into a DataFrame for easier manipulation:
  bedgraph_df = pd.read_csv(bedgraph_path, sep='\t', header=None, names=['chrom', 'start', 'end', 'value'])
  bedgraph_df

  # Select the C positions of interest identified in the 'C_CG_coords' vector
  selected_C_values = bedgraph_df[bedgraph_df['start'].isin(C_CG_coords_fwd)]
  selected_C_values = selected_C_values.sort_values(by=['start'], ascending=True)

  # Make sure that the order of the values is correct (aligned with the order set in 'C_CG_coords_fwd')
  has_all_sorted_Cs =sum(np.array(selected_C_values['start']) != np.array(C_CG_coords_fwd)) == 0
  print("Check that the values in the extracted mC vector are exactly the same as the ones identified in the Cs position list", has_all_sorted_Cs)

  mC_probs_per_position = list(selected_C_values['value'])
  print(mC_probs_per_position)
  return(mC_probs_per_position)
```

```python
bedgraph_path = "/content/Homo_sapiens_watson_CG_Cancer_Lymphoblastoid_SUB33-chr1-207320817..207320907.bedgraph"
is_forward_strand = False

extract_mC_probs_from_BS_bedgraph(bedgraph_path, is_forward_strand,
                                      C_CG_coords_fwd=np.array(C_CG_coords))


# "/content/Homo_sapiens_crick_CG_Pancreas_37yrs_Male-chr1-207320821..207320911.bedgraph"
# "/content/Homo_sapiens_crick_CG_HepG2_HepatocellularCarcinoma_15yrs_Male_child_Rep1-chr1-207320821..207320911.bedgraph"
# "/content/Homo_sapiens_watson_C_Tumor_HepG2_15yrs_Male_Rep2-chr1-207320824..207320905.bedgraph"
```
